## 5. Test Retrieval Playground Package

```python
print("Testing retrieval_playground package imports...\n")
print("=" * 60)

import os

# Check if required env vars exist (needed for some imports)
has_env_vars = (
    os.getenv("GOOGLE_API_KEY") and 
    os.getenv("QDRANT_URL") and 
    os.getenv("QDRANT_KEY")
)

if not has_env_vars:
    print("⚠️  Environment variables not set - some imports may fail")
    print("   (Set up .env file first for full verification)\n")

# Test imports from each module
test_imports = [
    # Core utilities - config is a module, not an attribute
    ("retrieval_playground.utils", "config", "Configuration (module)", True, "module"),
    ("retrieval_playground.utils.model_manager", "model_manager", "Model manager", False, "attribute"),
    
    # Pre-retrieval (Chunking)
    ("retrieval_playground.src.pre_retrieval.chunking.recursive_chunking", "RecursiveChunking", "Recursive chunking", False, "attribute"),
    ("retrieval_playground.src.pre_retrieval.chunking.contextual_chunking", "ContextualChunking", "Contextual chunking", False, "attribute"),
    ("retrieval_playground.src.pre_retrieval.chunking.parent_child_chunking", "ParentChildChunking", "Parent-child chunking", False, "attribute"),
    ("retrieval_playground.src.pre_retrieval.chunking.docling_chunking", "DoclingChunking", "Docling chunking", False, "attribute"),
    
    # Pre-retrieval (Query)
    ("retrieval_playground.src.pre_retrieval.query_rephrasing", "expand_query", "Query expansion", False, "attribute"),
    ("retrieval_playground.src.pre_retrieval.query_rephrasing", "optimize_query_for_retrieval", "Query optimization", False, "attribute"),
    ("retrieval_playground.src.pre_retrieval.routing", "semantic_layer", "Query routing", False, "attribute"),
    ("retrieval_playground.src.pre_retrieval.routing", "route_with_complexity_analysis", "Complexity analysis", False, "attribute"),
    
    # Mid-retrieval
    ("retrieval_playground.src.mid_retrieval.hybrid_search", "HybridRetriever", "Hybrid search", False, "attribute"),
    ("retrieval_playground.src.mid_retrieval.reranking", "Reranker", "Reranking", False, "attribute"),
    ("retrieval_playground.src.mid_retrieval.parent_child_retrieval", "ParentChildRetriever", "Parent-child retrieval", False, "attribute"),
    ("retrieval_playground.src.mid_retrieval.multi_query_hybrid", "MultiQueryHybrid", "Multi-query hybrid", False, "attribute"),
    ("retrieval_playground.src.mid_retrieval.route_driven_retrieval", "RouteRetriever", "Route-driven retrieval", False, "attribute"),
    ("retrieval_playground.src.mid_retrieval.adaptive_retrieval", "AdaptiveRetriever", "Adaptive retrieval", False, "attribute"),
    
    # Post-retrieval
    ("retrieval_playground.src.post_retrieval", "ContextPreparer", "Context preparation", False, "attribute"),
    ("retrieval_playground.src.post_retrieval", "document_assembly", "Document assembly", False, "attribute"),
    
    # Evaluation
    ("retrieval_playground.src.evaluation", "RAGEvaluator", "RAG evaluation", False, "attribute"),
    
    # Baseline RAG
    ("retrieval_playground.src.baseline_rag", "RAG", "Baseline RAG", False, "attribute"),
]

all_ok = True
skipped = []

for test in test_imports:
    if len(test) == 5:
        module_path, item_name, description, requires_env, import_type = test
    else:
        # Backwards compatibility
        module_path, item_name, description, requires_env = test
        import_type = "attribute"
    
    # Skip config if env vars not set
    if requires_env and not has_env_vars:
        print(f"⚠️  {description:30s} - SKIPPED (needs .env)")
        skipped.append(description)
        continue
    
    try:
        module = __import__(module_path, fromlist=[item_name])
        
        if import_type == "module":
            # For modules, just check if import succeeded
            if hasattr(module, item_name):
                getattr(module, item_name)
            print(f"✅ {description:30s} - imported")
        else:
            # For attributes, check they exist
            getattr(module, item_name)
            print(f"✅ {description:30s} - {item_name}")
    except Exception as e:
        # Show full error message for debugging
        error_msg = str(e)
        if len(error_msg) > 80:
            error_msg = error_msg[:80] + "..."
        print(f"❌ {description:30s} - FAILED")
        print(f"   Error: {error_msg}")
        all_ok = False

print("\n" + "="*60)
if all_ok and not skipped:
    print("✅ retrieval_playground package is properly installed!")
elif all_ok and skipped:
    print("⚠️  Partial check - some imports skipped (set up .env to test fully)")
    print(f"   Skipped: {', '.join(skipped)}")
else:
    print("❌ Some retrieval_playground modules failed to import")
    print("   Make sure you're running from the correct directory")
    print("   Try: pip install -e .")
```

## 6. Test API Connections (Optional)

```python
print("Testing API connections...\n")
print("=" * 60)
print("⚠️  This section makes actual API calls and may incur costs.")
print("   Skip this section if you want to avoid API charges.\n")

# Test Gemini API
try:
    from langchain_google_genai import GoogleGenerativeAI
    import os
    
    if os.getenv("GOOGLE_API_KEY"):
        llm = GoogleGenerativeAI(model="gemini-3.1-flash-lite", temperature=0)
        response = llm.invoke("Say 'Hello'")
        print("✅ Gemini API connection successful")
        print(f"   Response: {response[:50]}...")
    else:
        print("⚠️  Skipping Gemini test - API key not set")
except Exception as e:
    print(f"❌ Gemini API connection failed: {str(e)[:100]}")

print()

# Test Qdrant connection
try:
    from qdrant_client import QdrantClient
    import os
    
    if os.getenv("QDRANT_URL") and os.getenv("QDRANT_KEY"):
        client = QdrantClient(
            url=os.getenv("QDRANT_URL"),
            api_key=os.getenv("QDRANT_KEY")
        )
        collections = client.get_collections()
        print("✅ Qdrant connection successful")
        print(f"   Collections found: {len(collections.collections)}")
        
        # Check for required collections
        collection_names = [c.name for c in collections.collections]
        required = ["recursive_character", "hybrid", "parent_child"]
        found = [c for c in required if c in collection_names]
        
        if len(found) == len(required):
            print(f"   ✅ All required collections present: {', '.join(found)}")
        else:
            missing = [c for c in required if c not in collection_names]
            print(f"   ⚠️  Missing collections: {', '.join(missing)}")
            print(f"   ⚠️  Found collections: {', '.join(collection_names)}")
        
        client.close()
    else:
        print("⚠️  Skipping Qdrant test - credentials not set")
except Exception as e:
    print(f"❌ Qdrant connection failed: {str(e)[:100]}")

print("\n" + "="*60)
print("Note: API connection tests are optional")
print("      You can proceed even if these tests fail")
print("      Connections will be retested in the tutorial notebooks")
```
