## 2. Check Required Packages

```python
# Check all required imports
packages_to_check = [
    "langchain",
    "langchain_community",
    "langchain_google_genai",
    "langchain_huggingface",
    "langchain_qdrant",
    "qdrant_client",
    "sentence_transformers",
    "docling",
    "ragas",
    "pydantic",
    "pandas",
    "numpy",
]

print("Checking required packages...\n")
all_ok = True

for package in packages_to_check:
    try:
        __import__(package)
        print(f"✅ {package}")
    except ImportError:
        print(f"❌ {package} - NOT FOUND")
        all_ok = False

print("\n" + "="*50)
if all_ok:
    print("✅ All required packages are installed!")
else:
    print("❌ Some packages are missing. Please reinstall.")
```

## 3. Check Environment Variables

```python
import os
from dotenv import load_dotenv

# Load .env file
load_dotenv()

print("Checking environment variables...\n")

# Check API keys
google_key = os.getenv("GOOGLE_API_KEY")
openai_key = os.getenv("OPENAI_API_KEY")
qdrant_url = os.getenv("QDRANT_URL")
qdrant_key = os.getenv("QDRANT_KEY")

# Check LLM API keys (at least one required)
has_google = google_key and google_key != "your_google_api_key_here"
has_openai = openai_key and openai_key != "your_openai_api_key_here"

if has_google:
    masked = google_key[:8] + "..." + google_key[-4:]
    print(f"✅ Google Gemini API Key: {masked}")
else:
    print(f"⚠️  Google Gemini API Key: NOT SET")

if has_openai:
    masked = openai_key[:8] + "..." + openai_key[-4:]
    print(f"✅ OpenAI API Key: {masked}")
else:
    print(f"⚠️  OpenAI API Key: NOT SET")

# Check Qdrant credentials
has_qdrant_url = qdrant_url and qdrant_url != "your_qdrant_url_here"
has_qdrant_key = qdrant_key and qdrant_key != "your_qdrant_key_here"

if has_qdrant_url:
    print(f"✅ Qdrant URL: {qdrant_url}")
else:
    print(f"⚠️  Qdrant URL: NOT SET")

if has_qdrant_key:
    masked = qdrant_key[:8] + "..." + qdrant_key[-4:]
    print(f"✅ Qdrant API Key: {masked}")
else:
    print(f"⚠️  Qdrant API Key: NOT SET")

print("\n" + "="*50)
if (has_google or has_openai) and has_qdrant_url and has_qdrant_key:
    print("✅ All required environment variables are set!")
elif not (has_google or has_openai):
    print("⚠️  Missing LLM API key - set either GOOGLE_API_KEY or OPENAI_API_KEY")
    print("   (Required for running tutorial notebooks)")
else:
    print("⚠️  Missing Qdrant credentials")
    print("   (Required for running tutorial notebooks)")
```
